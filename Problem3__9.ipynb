```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Lorenz_attractor(x0, y0, z0, t):
  len_t = len(t)
  dt = t[1] - t[0]
  x = np.zeros(len_t)
  y = np.zeros(len_t)
  z = np.zeros(len_t)

  x[0], y[0], z[0] = x0, y0, z0

  for i in range(1, len_t-1):
    x[i] = x[i-1] + 10 * (y[i-1] - x[i-1]) * dt
    y[i] = y[i-1] + (-x[i-1]*z[i-1] + 28*x[i-1] - y[i-1]) * dt
    z[i] = z[i-1] + (x[i-1]*y[i-1] - 8/3 * z[i-1]) * dt

  return x, y, z
```

```python
t = np.linspace(0, 2, 100)
x, y, z = Lorenz_attractor(np.random.rand(), np.random.rand(), np.random.rand(), t)
```

```python
fig, ax =plt.subplots(3,1)
fig.set_figheight(15)
ax[0].set(xlabel = 'y', ylabel = 'z')
ax[0].grid()
ax[0].scatter(y, z, s=5)

ax[1].set(xlabel = 'x', ylabel = 'z')
ax[1].grid()
ax[1].scatter(x, z, s=5)

ax[2].set(xlabel = 'y', ylabel = 'x')
ax[2].grid()
ax[2].scatter(y, x, s=5)
```
