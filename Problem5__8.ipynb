```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def finite_diff(w1, w, a, alpha, beta, tf, n):

  h = tf/n
  r = np.zeros(n-1)
  t = np.linspace(0,tf, n)
  r = np.cos(w1*t)

  M = np.zeros((n-1, n-1))

  for i in range (n-1):
    for j in range (n-1):
        if i == j:
            M[i, j] = 2 + w**2 * h**2
        if j == i + 1:
            M[i, j] = a * h/2 - 1
        if i == j + 1:
            M[i, j] = -a * h/2 - 1

  B = np.zeros((n-1, 1))
  e0 = (-h/2 *a +1)*alpha
  eN = (-h/2 *a +1)*beta

  B[0,0] = -h**2 * r[0] + e0
  B[n-2,0] = -h**2 * r[n-2] + eN

  for i in range (1, n-3):
    B [i,0] = (-h) ** 2 * r[i]

  y = np.linalg.solve(M, B)

  return np.transpose(y)[0]
```

```python
y = finite_diff(1,1,1, 1, -.6, 10, 100)
plt.ylabel('y')
plt.xlabel('x')
plt.title('finite difference method')
plt.grid()
plt.plot(np.linspace(0,10, 99), y)
```
