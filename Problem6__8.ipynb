```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 1
N = 300
dx = L / (N - 1)  
iter = 100000


u_top = 1
u_bottom = np.linspace(0, L, N)   
u_left = np.linspace(0, L, N)**2   
u_right = np.zeros(N)  


u = np.zeros((N, N))
u[0, :] = u_top
u[-1, :] = u_bottom
u[:, 0] = u_left
u[:, -1] = u_right
   

for i in range(N):
  for j in range(N):
    r = np.random.random()
    u[i,j] = r


for i in range(iter):
      rx = np.random.randint(1,N-2)
      ry = np.random.randint(1,N-2)

      u[rx, ry] = (u[rx-1,ry] + u[rx+1,ry] + u[rx,ry-1] + u[rx,ry+1])/4
    
X, Y = np.meshgrid(np.linspace(0, L, N), np.linspace(0, L, N))
plt.contourf(X, Y, u)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```
