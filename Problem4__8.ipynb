```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def iterative_relaxation(y0, yf, tf, num_iter):

    n = num_iter
    y = np.zeros(n)
    y[0] = y0
    y[-1] = yf

    for i in range(1,n-1): 
      r = np.random.random()
      y[i] = r
    for j in range(1,n-1):
      r = np.random.randint(n-1)
      y[r] = (y[r-1] + y[r+1])/2

    return y

    
    
```

```python
y = iterative_relaxation(1, -0.60, 10, 100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relaxation method')
plt.grid()
plt.plot(np.linspace(0,10,100), y)
```
