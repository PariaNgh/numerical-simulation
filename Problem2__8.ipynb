```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f2(f):
  return f**2

def _f(f):
  return -f

def explicit(f0, x, deriv):
  n = len(x)
  f = np.zeros(n)
  f[0] = f0
  dx = x[1] - x[0]

  for i in range(1,n):
    f[i] = f[i-1] + dx * deriv(f[i-1])

  return f

def implicitf2(f0, x):
  n = len(x)
  f = np.zeros(n)
  f[0] = f0
  dx = x[1] - x[0]

  for i in range(1,n):
  
    f[i] = 1/(dx*2) + (1+4*dx*f[i-1])**0.5 / (dx*2)
  
  return f

def implicit_f(f0, x):
  n = len(x)
  f = np.zeros(n)
  f[0] = f0
  dx = x[1] - x[0]

  for i in range(1,n):
    f[i] = f[i-1] / (1 + dx)
  return f
```

# Solving f' = f^2 with explicit approach

```python
xf2 = np.linspace(0, 2, 5)
y_ex_f2 = explicit(1, xf2, f2)
plt.plot(xf2, y_ex_f2)
```

# Solving f' = f^2 with implicit approach

```python
y_im_f2 = implicitf2(1, xf2)
plt.plot(xf2, y_im_f2)
```

# Solving f' = -f with explicit approach

```python
x_f = np.linspace(0, 10, 21)
y_ex_f = explicit(1, x_f, _f)
plt.plot(x_f, y_ex_f)
```

# Solving f' = -f with implicit approach

```python
y_im_f = implicit_f(1, x_f)
plt.plot(x_f, y_im_f)
```
