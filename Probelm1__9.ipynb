```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def logistic_map(x0 ,miu):
  x = [x0]
  error = 0.01
  x_old, x_new = 0, 0
  i = 0
  for i in range(1000):
    x_new = x[i] * miu * (1 - x[i])
    x.append(x_new)

  return x[-1]
```

```python
num_miu = 500
miu = np.linspace(0,7, num_miu)
ens = 50
x0 = np.random.random(size = ens)
x_miu = np.zeros((ens,num_miu))

for i in range(num_miu):
  for e in range(ens):
   x_miu[e][i] = logistic_map(x0[e], miu[i])
  

```

```python
plt.figure(figsize = (20,5) )
plt.title('logistic map')
plt.xlabel('miu')
plt.ylabel('x')
plt.grid()
for i in range(ens):
  plt.plot(miu, x_miu[i])
```
